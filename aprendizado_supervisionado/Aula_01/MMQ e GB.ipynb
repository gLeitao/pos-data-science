```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
x = np.linspace(-100, 100, num=100)
y = x ** 2

x_linha = 100
alpha = 0.05

plt.plot(x, y, c='blue')
plt.plot([-100, 100], [0,0], 'k--')

for i in range(50):
    y_plot = x_linha ** 2
    plt.plot(x_linha, y_plot, 'o', c='red')
    
    x_linha -= alpha * (2 * x_linha)
    
```

```python
x = np.arange(200)
y = np.random.uniform(size=200) * 25 + range(200)

sxy = (x * y).sum() - (x.shape[0] * np.mean(x) * np.mean(y))
sxx = (x * y).sum() - (x.shape[0] * np.mean(x) * np.mean(x))
beta_1 = sxy / sxx
beta_0 = np.mean(y) - (beta_1 * np.mean(x))

y_hat = (x * beta_1) + beta_0

plt.scatter(x,y)
plt.plot(x, y_hat, c='red')
plt.show()

print(r2_score(y, y_hat))
print(np.sqrt(mean_squared_error(y, y_hat)))
```

```python
x = np.arange(200)
y = np.random.uniform(size=200) * 25 + range(200)

alpha = 0.00001
beta_0 = 9
beta_1 = 0

for i in range(25):
    y_hat = (x * beta_1) + beta_0
    
    plt.scatter(x,y)
    plt.plot(x, y_hat, c='red')
    plt.show()
    
    #atualizar os pesos em um passo do gradiente
    beta_0 -= alpha * -2 * np.mean((y - y_hat))
    beta_1 -= alpha *-2 * np.mean((y - y_hat) * x)
    
    print(r2_score(y, y_hat))
    print(np.sqrt(mean_squared_error(y, y_hat)))
    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python

x = np.linspace(0,10,200)
y = x * np.sin(x) + np.random.uniform(low=-1,high=1, size=200)

x = x.reshape(-1,1)
y = y.reshape(-1,1)

pf = PolynomialFeatures(degree=7, include_bias=False)
poly_x = pf.fit_transform(x)

lr = LinearRegression()
lr.fit(poly_x, y)
y_hat = lr.predict(poly_x)

plt.scatter(x,y)
plt.plot(x, y_hat, c='red')
plt.show()
```
